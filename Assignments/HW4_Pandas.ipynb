```python
MaxTemp=df["MaxTemp"].head(10)
AvgPressure=df["AvgPressure"].head(10)
plt.plot(MaxTemp, AvgPressure)
plt.show()
```

```python
MinTemp=df["MinTemp"].head(10)
AvgWind=df["AvgWind"].head(10)
AvgHumidity=df["AvgHumidity"].head(10)
```

```python
plt.plot(MaxTemp, AvgPressure)

plt.xlabel("MaxTemp")
plt.ylabel("AvgPressure")
plt.title("MaxTemp-AvgPressure Relation")

plt.show()
```

```python
plt.plot(MaxTemp, AvgPressure)

plt.xlabel("MaxTemp")
plt.ylabel("AvgPressure")
plt.title("MaxTemp-AvgPressure Relation")
plt.legend(["MaxTemp", "AvgPressure"],loc='upper right')

plt.show()
```

```python
plt.plot(MaxTemp, AvgPressure, color="b", linestyle="--")


plt.xlabel("MaxTemp")
plt.ylabel("AvgPressure")
plt.title("Temp-AvgPressure Relation")
plt.legend(["MaxTemp", "AvgPressure"])
plt.show()
```

```python
plt.plot(MaxTemp, AvgPressure,color="#1FC9A5", linestyle="--", label = "AvgPressure")
plt.plot(MaxTemp, AvgHumidity, color="#4EB772", label = "AvgHumidity")
plt.plot(MaxTemp, AvgWind, color="#A153E1", label = "AvgWind")



plt.xlabel("MaxTemp")
plt.ylabel("MinTemp")
plt.title("Temperature Relation")


plt.show()
```

```python
with plt.xkcd():
    plt.plot(MaxTemp, AvgHumidity, linewidth=3, label = "AvgHumidity")
    

    plt.xlabel("MaxTemp")
    plt.ylabel("AvgHumidity")
    plt.title("Temp-AvgHumidity Relation")
    
    plt.legend(loc='upper left')
    plt.grid(True)
    plt.tight_layout()
    plt.show() 
```

```python
plt.style.use("fivethirtyeight")
plt.bar(MaxTemp, AvgHumidity, linewidth=3, label = "AvgHumidity")
    

plt.xlabel("MaxTemp")
plt.ylabel("AvgHumidity")
plt.title("Temp-AvgHumidity Relation")
    
plt.legend(loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show() 
```

```python
plt.style.use("fivethirtyeight")

plt.title("Pie Chart")

labels = ["A","B","C","D","E","F","G","H","J","K"]
plt.pie(AvgHumidity, labels = labels)

plt.tight_layout()

plt.show( )
```
